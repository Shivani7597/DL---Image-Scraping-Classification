# Importing Necessary Libraries

```python
#Importing all required libraries for web scrapping
import selenium
import pandas as pd
import time
from selenium import webdriver
import requests
import urllib.request
import numpy as np
from bs4 import BeautifulSoup
from selenium.common.exceptions import StaleElementReferenceException, NoSuchElementException
```

# Scrapping Sarees from Amazon

```python
#Activating the chrome browser
driver=webdriver.Chrome(r"C:\Users\MUNGI\Downloads\chromedriver_win32 (1)\chromedriver.exe")

#Opening homepage of amazon.com
driver.get('https://www.amazon.in/')
#Asking the user to input the keywords to search
user_inp=input('Enter the product you want to search : ')
```

```python
#Searching of products using search tab and search button
search_bar=driver.find_element_by_id("twotabsearchtextbox")   #Locating search_bar by id
search_bar.clear()                                            #Clearing search bar
search_bar.send_keys(user_inp)                                #sending user input to search bar
time.sleep(4)
search_button=driver.find_element_by_xpath('//div[@class="nav-search-submit nav-sprite"]')   #Locating search bar by id
search_button.click()
```

```python
#Creating empty list to store the urls of images
img_urls=[]
```

```python
#Fetching urls of images from the first 5 pages 
for i in range(0,5):
    for j in driver.find_elements_by_xpath("//div[@class='a-section aok-relative s-image-tall-aspect']/img"):   #path of images
        img_urls.append(j.get_attribute("src"))
        driver.find_element_by_xpath("//li[@class='a-last']/a")    #clicking on the next button
        time.sleep(3)
```

```python
#urls of the images
img_urls
```

```python
#Length of the image urls stored in the list
len(img_urls)
```

```python
#Downloading images in the required folder from the urls obtained
for i in range(len(img_urls)):
    if i >= 300:
        break
    print("Downloading {0} of {1} images".format(i, 300))
    response= requests.get(img_urls[i])
    file = open(r"M:\Flip Robo Internship\Deep Learning\Solution\Data Directory\Sarees\Saree"+str(i)+".jpg", "wb")
    file.write(response.content)
```

```python
#Closing current driver
driver.close()
```
